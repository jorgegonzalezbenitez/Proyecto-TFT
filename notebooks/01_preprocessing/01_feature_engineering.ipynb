Sanity check

```python
paths = glob.glob(str(OUT_DIR / "*.parquet"))
print("Parquets generados:", paths)
 
for p in paths:
    df      = pd.read_parquet(p)
    rn_cols = [c for c in df.columns if c.startswith("rn_")]
 
    print(f"\n📁 {Path(p).name}")
    print("shape:", df.shape)
    print("nº cols de canal:", len(rn_cols))
 
    required = {"fecha", "hotel", "ocup_total", "is_weekend", "season"}
    faltan   = required - set(df.columns)
 
    if faltan:
        print("❌ Faltan columnas:", faltan)
    else:
        print("✅ Base hotel OK")
 
    display(df[["fecha", "hotel", "ocup_total"] + rn_cols[:5]].head(3))
```
