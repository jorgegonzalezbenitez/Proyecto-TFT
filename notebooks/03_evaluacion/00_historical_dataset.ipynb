Imports y Path

```python
from pathlib import Path
import pandas as pd

BASE       = Path("../../")
DIR        = BASE / "data" / "features" / "hotel_ttoo"
OUTPUT_DIR = BASE / "dashboard" / "data"
OUTPUT_DIR.mkdir(parents=True, exist_ok=True)
print("Data dir:",   DIR)
print("Output dir:", OUTPUT_DIR)
```

Cargar y Estandarizar

```python
files = {
    "HOTEL_1": DIR / "HOTEL_1.parquet",
    "HOTEL_2": DIR / "HOTEL_2.parquet",
    "HOTEL_3": DIR / "HOTEL_3.parquet",
}
dfs = []
for hotel, path in files.items():
    print(f"Cargando {hotel} desde {path} ...")
    df = pd.read_parquet(path)
    df["hotel"] = hotel
    df["fecha"] = pd.to_datetime(df["fecha"], unit="ms")
    df.columns  = df.columns.str.lower()
    dfs.append(df)
```

Unir los 3 hoteles

```python
historical_df = pd.concat(dfs, ignore_index=True)
historical_df = historical_df.sort_values(["hotel", "fecha"]).reset_index(drop=True)
print("Tamaño final:", historical_df.shape)
historical_df.head()
```

Exportar

```python
out_file = OUTPUT_DIR / "historical_dataset.parquet"
historical_df.to_parquet(out_file, index=False)
print("✅ historical_dataset guardado en:", out_file)
```
